```python

# minx = 150.300861
# miny = -34.423526
# maxx = 151.589260
# maxy = -33.227966

minx = 150.301560
miny = -34.316218
maxx = 151.477327
maxy = -33.403931

crs = "EPSG:4326"
```

```python
geom = box(minx, miny, maxx, maxy)

bbox = gpd.GeoDataFrame(
    geometry=[geom],
    crs=crs
)
```

```python
bbox_reproj = bbox.to_crs("EPSG:3577")

bbox_reproj.to_file("bbox_sm.gpkg", driver="GPKG")
```

```python
area = bbox_reproj.area[0]
bounds = bbox_reproj.bounds.loc[0]
width = bounds.maxx - bounds.minx
height = bounds.maxy - bounds.miny

print(area/ 1e6)
print(height)
print(width)
```
